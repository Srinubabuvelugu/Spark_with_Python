```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("DF_to_CSV.com").getOrCreate()
```

```python
data=[(1,"sreenu"), (2,"Babu"),(3,"Siva")]
df = spark.createDataFrame(data=data,schema=['id','name'])
print(df.printSchema())
print(df.show())
```
