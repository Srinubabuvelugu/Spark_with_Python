```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.master('local[5]').appName('Repartion_Vs_Coalesec.com').getOrCreate()
```

```python
spark.sparkContext.master
```

```python
spark.sparkContext.appName
```

```python
## creating rdd using parallelize
rdd = spark.sparkContext.parallelize(range(20))
print('Partition size: ', rdd.getNumPartitions())

```

```python
rdd.savaAsTextFile("/temp/partitions")
```

```python
## creating rdd using parallelize with partition
rdd2 = spark.sparkContext.parallelize(range(20),6)
print('Partition size: ', rdd2.getNumPartitions())
```

```python
## Increasing the partitions using repartition 
rdd_repartition = rdd.repartition(8)
print("rdd_repartition: ",rdd_repartition.getNumPartitions())
```

```python
## ## Increasing the partitions using repartition 
rdd_repartition_dec = rdd.repartition(3)
print("rdd_repartition_dec: ",rdd_repartition_dec.getNumPartitions())
```

```python
## Coalesce is a transformation method on rdd, it is used to reduce the partition size without full shuffle of data in rdd
## Coalesec
rdd_coalesce = rdd.coalesce(4)
print("rdd_repartition with coalesec: ",rdd_coalesce.getNumPartitions())
```

```python
## Coalesce is a transformation method on rdd, it is used to reduce the partition size without full shuffle of data in rdd
## Coalesec
## If we try to increase the partition size, it will not happen,
## it returns the original partition size
rdd_coalesce_inc = rdd.coalesce(6)
print("rdd_repartition with coalesec: ",rdd_coalesce_inc.getNumPartitions())
```

```python
# DataFrame example
df=spark.range(0,20)
print(df.rdd.getNumPartitions())

df.write.mode("overwrite").csv("c:/tmp/partition.csv")
```
